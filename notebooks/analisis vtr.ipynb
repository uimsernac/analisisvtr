```python
import pandas as pd
```

```python
base2020=pd.read_excel(r"G:\Unidades compartidas\UIM\2024\002 Proyectos\042 analisis vtr\analisisvtr\Datos\Reclamos vtr 2020.xlsx")
base2019=pd.read_excel(r"G:\Unidades compartidas\UIM\2024\002 Proyectos\042 analisis vtr\analisisvtr\Datos\Reclamos vtr Octubre-Diciembre 2019.xlsx")
```

```python
combined_df = pd.concat([base2019, base2020], ignore_index=True)
```

```python
combined_df
```

```python
combined_df['Tipo de Motivo'].value_counts()
```

```python
# Convert the 'Tipo de Motivo' column to uppercase
combined_df['Tipo de Motivo'] = combined_df['Tipo de Motivo'].str.upper()

# Define the filter criteria
filter_criteria = ["CONTINUIDAD DE SERVICIO", "CONEXIONES A INTERNET", "VELOCIDAD DE INTERNET"]

# Filter the DataFrame
filtered_df = combined_df[combined_df['Tipo de Motivo'].isin(filter_criteria)]
filtered_df
```

```python
# Count the occurrences of each unique value in the 'RUT' column
rut_counts = filtered_df['Rut'].value_counts().reset_index()

# Rename the columns for clarity
rut_counts.columns = ['Rut', 'Count']
```
